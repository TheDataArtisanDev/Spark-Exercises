```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("IcebergTimeTravel") \
    .config("spark.sql.catalog.local", "org.apache.iceberg.spark.SparkCatalog") \
    .config("spark.sql.catalog.local.type", "hadoop") \
    .config("spark.sql.catalog.local.warehouse", "/tmp/iceberg_warehouse") \
    .config("spark.jars.packages", "org.apache.iceberg:iceberg-spark-runtime-3.5_2.12:1.4.2") \
    .getOrCreate()
```

```python
# View snapshot history
spark.sql("SELECT * FROM local.db.sample_users.snapshots").show(truncate=False)
```

```python
# View manifest files for each snapshot
spark.sql("SELECT * FROM local.db.sample_users.manifests").show(truncate=False)
```

```python
# View logical data files in use
spark.sql("SELECT * FROM local.db.sample_users.files").show(truncate=False)
```
